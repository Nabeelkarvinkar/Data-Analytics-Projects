```python
#Using CASE WHEN to create buckets for age by adding a column age_bucket 
ps.sqldf("""

SELECT  Name, PassengerId, Age,
    CASE WHEN Age IS NULL THEN '0. Missing'
        WHEN Age <18 THEN '1. 1-17'
        WHEN Age <60 THEN '2. 18-60'
    ELSE '3. 60+' END AS age_bucket
FROM  df
LIMIT 50   
    
""")
```

```python
#Getting all distinct values of "Embarked" column (UNIQUE VALUES)
ps.sqldf("""

SELECT 
    DISTINCT Embarked FROM df
    
""")
```

```python
#Aggregating information at "Embarked" column level (group by -embarked)
ps.sqldf("""

SELECT Embarked,
       AVG(Age) AS AVG_AGE,
       SUM(Age) AS TOTAL_AGE,
       MIN(Age) AS MIN_AGE,
       MAX(Age) AS MAX_AGE,
       COUNT(PassengerId) AS TOTAL_PASSENGER,
       COUNT(DISTINCT Ticket) AS DISTN_TICKETS
       FROM df 
       GROUP BY Embarked

""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels
ps.sqldf("""

SELECT Embarked,Sex,
       AVG(Age) AS AVG_AGE,
       SUM(Age) AS TOTAL_AGE,
       MIN(Age) AS MIN_AGE,
       MAX(Age) AS MAX_AGE,
       COUNT(PassengerId) AS TOTAL_PASSENGER,
       COUNT(DISTINCT Ticket) AS DISTN_TICKETS
       FROM df 
       GROUP BY Embarked, Sex

""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels (groupby -Embarked and Sex)
#and also using "WHERE" to filter basis some conditions
ps.sqldf("""

SELECT Embarked,Sex,
       AVG(Age) AS AVG_AGE,
       SUM(Age) AS TOTAL_AGE,
       MIN(Age) AS MIN_AGE,
       MAX(Age) AS MAX_AGE,
       COUNT(PassengerId) AS TOTAL_PASSENGER,
       COUNT(DISTINCT Ticket) AS DISTN_TICKETS
       FROM df 
       WHERE Age>=18 AND Age<=60
       GROUP BY Embarked, Sex

""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also sorting the output by order by Name(By default it is ascending, for descending DESC).
ps.sqldf("""

SELECT Embarked,Name,Sex,
       AVG(Age) AS AVG_AGE,
       SUM(Age) AS TOTAL_AGE,
       MIN(Age) AS MIN_AGE,
       MAX(Age) AS MAX_AGE,
       COUNT(PassengerId) AS TOTAL_PASSENGER,
       COUNT(DISTINCT Ticket) AS DISTN_TICKETS
       FROM df 
       WHERE Age>=18 AND Age<=60
       GROUP BY Embarked, Sex
       ORDER BY Name

""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also using "HAVING" for fitering the output for buckets where average age is more than 30
#and also using "ORDER BY" for sorting the output in descending order
ps.sqldf("""

SELECT Embarked,Name,Sex,
       AVG(Age) AS AVG_AGE,
       SUM(Age) AS TOTAL_AGE,
       MIN(Age) AS MIN_AGE,
       MAX(Age) AS MAX_AGE,
       COUNT(PassengerId) AS TOTAL_PASSENGER,
       COUNT(DISTINCT Ticket) AS DISTN_TICKETS
       FROM df 
       WHERE Age>=18 AND Age<=60
       GROUP BY Embarked, Sex
       HAVING AVG_AGE> 30
       ORDER BY Name

""")
```

```python
cities = pd.DataFrame({'code':['S', 'C', 'L'],'city':['Southampton', 'Cherbourg', 'London']})
cities
```

```python
#Inner join between cloumn Embarked and Code from Cites data.(only common data )
ps.sqldf("""

SELECT Embarked,Name,Sex,
       AVG(Age) AS AVG_AGE,
       SUM(Age) AS TOTAL_AGE,
       MIN(Age) AS MIN_AGE,
       MAX(Age) AS MAX_AGE,
       COUNT(PassengerId) AS TOTAL_PASSENGER,
       COUNT(DISTINCT Ticket) AS DISTN_TICKETS
       FROM df
       
       INNER JOIN cities C
       ON Embarked=C.code
       
       GROUP BY Embarked, Sex
       ORDER BY Name

""")
```

```python
#Left join between cloumn Embarked and Code from Cites data.
#(common data from both column and also remaining data from left column ie.Embarked )
ps.sqldf("""

SELECT Embarked,Name,Sex,
       AVG(Age) AS AVG_AGE,
       SUM(Age) AS TOTAL_AGE,
       MIN(Age) AS MIN_AGE,
       MAX(Age) AS MAX_AGE,
       COUNT(PassengerId) AS TOTAL_PASSENGER,
       COUNT(DISTINCT Ticket) AS DISTN_TICKETS
       FROM df
       
       LEFT JOIN cities C
       ON Embarked=C.code
       
       GROUP BY Embarked, Sex
       ORDER BY Name

""")
```

```python
#RANK() sorts the data, gives the same rank to all consecutive n rows with same values, 
#but for next row with a different value, it gives a rank equal to previous rank plus n
ps.sqldf("""

SELECT  Name, Embarked,Sex,
        Age, Fare,
        RANK() OVER(ORDER BY Age DESC) AS rnk
        FROM  df
        LIMIT 10

""")
```

```python
#DENSE_RANK() does not skip ranks like RANK() does
ps.sqldf("""

SELECT  Name, Embarked,Sex,
        Age, Fare,
        DENSE_RANK() 
        OVER(ORDER BY Age DESC) AS rnk
        FROM  df
        LIMIT 10

""")
```

```python
#ROW_NUMBER() sorts the data and gives different ranks to each row 
ps.sqldf("""

SELECT  Name, Embarked,Sex,
        Age, Fare,
        ROW_NUMBER() 
        OVER(ORDER BY Age DESC) AS rnk
        FROM  df
        LIMIT 10

""")
```

```python
#You can use multiple columns to sort the data within OVER() function
ps.sqldf("""

SELECT  Name, Embarked,Sex,
        Age, Fare,
        ROW_NUMBER() 
        OVER(ORDER BY Age DESC, Fare DESC) AS rnk
        FROM  df
        LIMIT 10

""")
```

```python
#Using PARTITION BY to create buckets for giving separate sets of ranks
ps.sqldf("""

SELECT  Name, Embarked,Sex,
        Age, Fare,
        ROW_NUMBER() 
        OVER(PARTITION BY Embarked,Sex ORDER BY Age DESC, Fare DESC) AS rnk
        FROM  df
        LIMIT 10


""")
```

```python
#Examples for LEAD() and LAG()
ps.sqldf("""

SELECT 
    Name,
    LEAD(Name, 1) OVER(ORDER BY Name) AS next_name,
    LEAD(Name, 2) OVER(ORDER BY Name) AS second_next_name,
    LAG(Name, 1) OVER(ORDER BY Name) AS previous_name,
    LAG(Name, 2) OVER(ORDER BY Name) AS second_previous_name
    
FROM df
LIMIT 5

""")
```

```python
#Using subqueries to get average age and then filtering only those rows where average age is below 30
ps.sqldf("""

SELECT Embarked, avg_age
    
FROM (SELECT  Embarked, AVG(Age) AS avg_age  FROM df 
      GROUP BY Embarked) AS df1

WHERE df1.avg_age<30
    
""")
```

```python
#Using subqueries to get average age and number of passengers separately and then joining the outputs
ps.sqldf("""

SELECT 
    df1.Embarked,
    df1.avg_age,
    df2.tot_passengers
FROM (SELECT 
            Embarked, 
            AVG(Age) AS avg_age 
        FROM df 
        GROUP BY Embarked) AS df1

INNER JOIN (SELECT 
                Embarked, 
                COUNT(PassengerID) AS tot_passengers
            FROM DF
            GROUP BY Embarked) AS df2
    ON df1.Embarked = df2.Embarked
    
""")
```

```python
#Using nested subqueries to filter data
ps.sqldf("""

SELECT * FROM df
WHERE Embarked IN 
       (SELECT  DISTINCT df1.Embarked
        FROM (SELECT Embarked, AVG(Age) AS avg_age 
                FROM df
                GROUP BY Embarked) AS df1
        WHERE df1.avg_age>30)
    
""")
```
